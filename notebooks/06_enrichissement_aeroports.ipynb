# 🗺️ 06 - Enrichissement avec infos aéroports

Ajout des noms, pays et villes d'origine et de destination à chaque vol.

```python
import pandas as pd

# Charger les données de vols
df = pd.read_csv("../data/flights_sample_3m.csv")

# Charger les données d'aéroports
airports = pd.read_csv("../data/airports.csv")
airports = airports[['ident', 'name', 'iso_country', 'municipality', 'latitude_deg', 'longitude_deg']]
airports.columns = ['AIRPORT', 'NAME', 'COUNTRY', 'CITY', 'LAT', 'LON']
```

## 🔁 Fusion avec ORIGIN

```python
# Fusion avec aéroport d'origine
df = df.merge(airports, how='left', left_on='ORIGIN', right_on='AIRPORT')
df = df.rename(columns={
    'NAME': 'ORIGIN_NAME',
    'COUNTRY': 'ORIGIN_COUNTRY',
    'CITY': 'ORIGIN_CITY',
    'LAT': 'ORIGIN_LAT',
    'LON': 'ORIGIN_LON'
})
df = df.drop('AIRPORT', axis=1)
```

## 🔁 Fusion avec DEST

```python
# Fusion avec aéroport de destination
df = df.merge(airports, how='left', left_on='DEST', right_on='AIRPORT')
df = df.rename(columns={
    'NAME': 'DEST_NAME',
    'COUNTRY': 'DEST_COUNTRY',
    'CITY': 'DEST_CITY',
    'LAT': 'DEST_LAT',
    'LON': 'DEST_LON'
})
df = df.drop('AIRPORT', axis=1)
```
