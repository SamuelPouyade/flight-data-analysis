# ‚úàÔ∏è 02 - Analyse des retards

Exploration des causes de retard, des a√©roports concern√©s, et premi√®res visualisations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Affichage large
pd.set_option('display.max_columns', None)

# Charger les donn√©es
df = pd.read_csv("../data/flights_sample_3m.csv")
```

## üîç Moyenne des retards par cause

```python
delay_cols = [
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

df[delay_cols].mean().sort_values(ascending=False).plot(kind='bar', title="Retard moyen par cause")
plt.ylabel("Minutes")
plt.tight_layout()
plt.show()
```

## üõ´ Top 10 a√©roports de d√©part

```python
df['ORIGIN'].value_counts().head(10).plot(kind='barh', title="Top 10 a√©roports de d√©part")
plt.xlabel("Nombre de vols")
plt.tight_layout()
plt.show()
```

## ‚ùå R√©partition des vols annul√©s

```python
df['CANCELLED'].value_counts(normalize=True).plot(kind='pie', labels=['Non annul√©', 'Annul√©'], autopct='%1.1f%%')
plt.title("Part des vols annul√©s")
plt.ylabel("")
plt.show()
```
