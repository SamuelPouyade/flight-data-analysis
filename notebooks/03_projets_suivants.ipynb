## 📈 Corrélation entre types de retards

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/flights_sample_3m.csv")
delay_cols = [
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

sns.heatmap(df[delay_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Corrélations entre types de retards")
plt.show()
```

## ⏱️ Retards selon l'heure de départ

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
df = pd.read_csv("../data/flights_sample_3m.csv")

df['SCHEDULED_DEPARTURE_HOUR'] = df['SCHED_DEP'] // 100
df.groupby('SCHEDULED_DEPARTURE_HOUR')['DEP_DELAY'].mean().plot(kind='bar')
plt.title("Retard moyen par heure de départ")
plt.xlabel("Heure prévue de départ")
plt.ylabel("Retard moyen (minutes)")
plt.tight_layout()
plt.show()
```

## 📏 Distance vs Retard

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
df = pd.read_csv("../data/flights_sample_3m.csv")
sns.scatterplot(data=df, x='DISTANCE', y='DEP_DELAY', alpha=0.3)
plt.title("Relation entre distance et retard")
plt.xlabel("Distance (miles)")
plt.ylabel("Retard au départ (minutes)")
plt.grid(True)
plt.show()
```

## ⚠️ Détection des retards extrêmes
